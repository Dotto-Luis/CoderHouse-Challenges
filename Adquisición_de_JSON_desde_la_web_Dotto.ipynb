# Adquisición de JSON desde la web


*   Escribir un código que lea un JSON desde esta esta [web api ](https://jsonplaceholder.typicode.com/users)
*   Pasarlo a un DataFrame




```python
import requests
import json
import pandas as pd
```

```python
url_base = 'https://jsonplaceholder.typicode.com/users'
url = requests.get(url_base)
```

```python
json_text = url.text
json.loads(json_text)
```

```python
json_data=json.loads(json_text)
df = pd.DataFrame.from_dict(json_data)
df.head()
```
